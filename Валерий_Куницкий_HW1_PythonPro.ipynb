# **Задание 1.**

Студент магистратуры обычно спит ночью XX часов и устраивает себе днем тихий час на YY минут. Определите, сколько всего студент спит в сутки (если вообще спит). Программа должна принимать значения XX и YY из стандартного потока ввода (функция input), результат надо выводить в стандартный поток вывода (функция print).


```
Sample Input 1:
7
30
Sample Output 1:
450

Sample Input 2:
0
42
Sample Output 2:
42
```


````python
#Решение задания 1
#YOUR CODE HERE

def run_z1():
  '''
  Вычисляем время сна в сутки

  Значения переменных:
  nh = часов ночного сна
  dm = минут дневного сна
  sm = сумма минут сна за сутки
  '''
  nh = input('Сколько часов вы спите ночью? ')
  try:
    nh = int(nh)
    if nh<0 or nh>24: return print('Исчерпан запас времени в сутках')
    else: dm = input('Сколько минут вы спите днём? ')
    
    dm = int(dm)
    if dm <0 or dm>(24*60 - nh*60):
      return print('Исчерпан запас времени в сутках')
    else: sm = nh*60 + dm

    if sm > 0: print('За сутки вы спите:',\
                     sm//60, 'часов,', sm%60, 'минут. (или',sm,'минут)')
    else: print('Вы вообще не спите')

  except:
    return print('Некорректный ввод')
  
  return 'Завершено успешно'


for i in range(7):
  print('\n\nРешение задания № 1')
  run_z1()
````

# **Задание 2**

Студент МФТИ каждый день ложится спать ровно в полночь и недавно узнал, что оптимальное время для его сна составляет XX минут. Он хочет поставить себе будильник так, чтобы он прозвенел ровно через XX минут после полуночи, однако для этого необходимо указать время сигнала в формате часы, минуты. ```

0

Sample Input 2:
512

````python
#Решение задания 2
#YOUR CODE HERE

def run_z2():
  '''
  Вычисляем время установки будильника
  
  Значения переменных:
  to_bed_time = время засыпания, выраженное в количестве
                минут от полуночи
  opt_sleep_duration = время оптимального сна в минутах
  alarm_time = время установки будильника, выраженное в количестве
               минут от полуночи
  '''
  disclaimer = '\n'+'Учитывайте текущее время суток в момент установки будильника!\n'+\
  'Значение времени будильника должно быть меньше текущего времени суток в цикле между 00:00 и 23:59!'

  to_bed_time = input('Вы ложитесь в полночь, верно? (y/n): ')
  if to_bed_time == 'y': to_bed_time = 0
  elif to_bed_time == 'n':
    to_bed_time = input('А во сколько? (введите <чч:мм> в формате 24часа, например 22:20): ')
    try:
      if int(to_bed_time[:2]) > 23: return print('Некорректное время')
      elif int(to_bed_time[3:5]) > 59: return print('Некорректное время')
      else: to_bed_time = int(to_bed_time[:2])*60 + int(to_bed_time[3:5])
    except: return print('Некорректный ввод')
  else: return print('Некорректный ответ')

  opt_sleep_duration = input('Сколько минут вам рекомендовали спать? ')
  try:
    opt_sleep_duration = int(opt_sleep_duration)
    if opt_sleep_duration >= 24*60:
      return print('Максимально поддерживаемая длительность составляет ', 24*60-1, ' минут')
  except: return print('Некорректный ввод')

  alarm_time = (to_bed_time + opt_sleep_duration) % (24*60)
  
  print('Замечательно, установите будильник на:')
  print(alarm_time//60, ' ч.')
  print(alarm_time%60, ' мин.')
  print(disclaimer)
  
  return 'Завершено успешно'


for i in range(6):
  print('\n\nРешение задания № 2')
  run_z2()
````

# **Задание 4**

Напишите программу, на вход которой подается список чисел одной строкой. Программа должна для каждого элемента этого списка вывести сумму двух его соседей. Для элементов списка, являющихся крайними, одним из соседей считается элемент, находящийся на противоположном конце этого списка. Например, если на вход подаётся список "1 3 5 6 10", то на выход ожидается список "13 6 9 15 7" (без кавычек).

Если на вход пришло только одно число, надо вывести его же.
Вывод должен содержать одну строку с числами нового списка, разделенными пробелом.


```
Sample Input 1:
1 3 5 6 10
Sample Output 1:
13 6 9 15 7
Sample Input 2:
10
Sample Output 2:
10
```


````python
#Решение задания 4
#YOUR CODE HERE

def run_z4():
  print('Введите последовательность чисел через пробел, и программа')
  print('посчитает сумму соседних элементов для каждого элемента:')
  input_list = list(map(int, input().split()))
  '''
  Здесь полученную из стандартного ввода строку (str)
  методом split() превращаем в список (list()) элементов str,
  затем с помощью map() преобразовываем тип элементов списка в числа (int),
  и далее из объекта "map" получаем список (list()) из int элементов.
  '''

  i = 0
  output_list = ''
  length = len(input_list)

  # Если пришло одно число:
  if length == 1:
    output_list = input_list[0]
  # Если пришло больше одного числа:
  else:
    for i in range(length):
      # Операции с соседями, пока элемент не последний.
      # Для первого элемента индекс соседа слева будет равен -1,
      # что и вернёт последний элемент списка
      if i < length - 1:
        output_list += str(input_list[i-1]+input_list[i+1]) + ' '
      # операции с соседями, для последнего элемента
      else:
        output_list += str(input_list[i-1]+input_list[0])

  print(output_list)


for i in range(3):
  print('\n\nРешение задания № 4')
  run_z4()
````

# **Задание 5**

Напишите программу, которая принимает на вход список чисел в одной строке и выводит на экран в одну строку значения, которые встречаются в нём более одного раза.

Для решения задачи может пригодиться метод sort списка (а может и не пригодиться).

Выводимые числа не должны повторяться, порядок их вывода может быть произвольным. Разделитель чисел - пробел


````python
#Решение задания 5
#YOUR CODE HERE

def run_z5():
  # Обработка ввода аналогично Заданию 4
  input_list = list(map(int, input().split()))

  repeat_list = set()

  for el, i in zip(input_list, range(len(input_list))):
    i2 = 0
    for i2 in range(len(input_list)):
      if i != i2:
        if input_list[i2] == el:
          repeat_list.add(el)
          break

  # Преобразуем множество элементов из чисел
  # в строку из чисел, разделённых пробелами
  output = ''
  for el in repeat_list: output += str(el) + ' '
  output = output[:-1] # отрезаем лишний пробел в конце

  print(output)


for i in range(3):
  print('\n\nРешение задания № 5')
  run_z5()
````

# **Задание 6**

GC-состав является важной характеристикой геномных последовательностей и определяется как процентное соотношение суммы всех гуанинов и цитозинов к общему числу нуклеиновых оснований в геномной последовательности.

Напишите программу, которая вычисляет процентное содержание символов G (гуанин) и C (цитозин) в введенной строке (программа не должна зависеть от регистра вводимых символов).
Например, в строке "acggtgttat" процентное содержание символов G и C равно 4*0.1 * 100 = 40.0, где 4 - это количество символов G и C, а 10 - это длина строки.

```
Sample Input:
acggtgttat
Sample Output:
40.0
```


````python
#Решение задания 6
#YOUR CODE HERE

def run_z6():
  chain = input()
  gc = 0

  for el in chain:
    if el.lower() == 'g' or el.lower() == 'c':
      gc += 1

  p = 100*gc/len(chain)

  print(p)


for i in range(3):
  print('\n\nРешение задания № 6')
  run_z6()
````

# **Задание 7**

Узнав, что ДНК не является случайной строкой, только что поступившие в МФТИ студенты группы аналитиков данных предложили использовать алгоритм сжатия, который сжимает повторяющиеся символы в строке.
Кодирование осуществляется следующим образом:
s = 'aaaabbсaa' преобразуется в 'a4b2с1a2', то есть группы одинаковых символов исходной строки заменяются на этот символ и количество его повторений в этой позиции строки.

Напишите программу, которая считывает строку, кодирует её предложенным алгоритмом и выводит закодированную последовательность на стандартный вывод. Кодирование должно учитывать регистр символов.


```
Sample Input 1:
aaaabbcaa
Sample Output 1:
a4b2c1a2
Sample Input 2:
abc
Sample Output 2:
a1b1c1
Sample Input 3:
AaAaBBBbbbCCc
Sample Output 3:
A1a1A1a1B3b3C2c1
```


````python
#Решение задания 7
#YOUR CODE HERE

def run_z7():
  chain = input()
  symbol = chain[0]
  index = 1
  eli = 2
  zipchain = ''

  for el in chain[1:]:
    if el == symbol:
      index += 1
    if el != symbol:
      zipchain += symbol + str(index)
      symbol = el
      index = 1
    if eli == len(chain):
      zipchain += symbol + str(index)
    eli += 1

  print(zipchain)


for i in range(3):
  print('\n\nРешение задания № 7')
  run_z7()
````

# **Задание 8**

Напишите программу, которая считывает с консоли числа (по одному в строке) до тех пор, пока сумма введённых чисел не будет равна 0 и сразу после этого выводит сумму квадратов всех считанных чисел.
Гарантируется, что в какой-то момент сумма введённых чисел окажется равной 0, после этого считывание продолжать не нужно.
В примере мы считываем числа 1, -3, 5, -6, -10, 13; в этот момент замечаем, что сумма этих чисел равна нулю и выводим сумму их квадратов, не обращая внимания на то, что остались ещё не прочитанные значения.


```
Sample Input:
1
-3
5
-6
-10
13
4
-8
Sample Output:
340
```

````python
#Решение задания 8
#YOUR CODE HERE

def run_z8():
  in_dig_list = []
  sumsqr = 0

  while True:
    in_dig = int(input())
    in_dig_list.append(in_dig)
    sumsqr += in_dig**2
    if sum(in_dig_list) == 0:
      print('\n', sumsqr)
      break


for i in range(3):
  print('\n\nРешение задания № 8')
  run_z8()
````

# **Задание 9**

Самое время использовать математику. Вы поднимаетесь по лестнице. Количество ступенек в ней = n.

На каждом этапе вы можете переступить либо одну, либо две ступеньки. Как много уникальных способов добраться до верха существует?

*1<=n<=45*



```
Input: n = 2
Output: 2
Explanation: There are two ways to climb to the top.
1. 1 step + 1 step
2. 2 steps

Input: n = 3
Output: 3
Explanation: There are three ways to climb to the top.
1. 1 step + 1 step + 1 step
2. 1 step + 2 steps
3. 2 steps + 1 step
```



````python
#Решение задания 9
#YOUR CODE HERE

def run_z9():
  '''
  Задача сводится к вычислению следующей задачи комбинаторики:
  "сколькими способами можно выбрать m из n различных предметов?"
  Для этого существует формула n!/(m!(n-m)!)

  Где для нашего случая <m> - количество двухступенчатых шагов.
  Стало быть будем перебирать от 0 до максимально возможного
  количества двухступенчатых шагов для заданного числа ступеней.
  А <n> в нашем случе - общее число шагов, необходимых,
  чтобы преодолеть лестницу, причём это конфигурация из
  одноступенчатых и/или двухступенчатых шагов.

  <n> в нашем случае меняется в зависимости от <m>
  для каждого варианта, что мы и переберём в одном цикле.
  '''
  import math

  try:
    stair_length = int(input('Сколько ступенек у лестницы? '))
    if stair_length > 45: raise ValueError

    max_2steps = stair_length//2
    variants_count = 0

    for count_of_2steps in range(max_2steps+1):
      # Зададим коэффициенты m, n для формулы из комбинаторики
      m = count_of_2steps
      n = stair_length - count_of_2steps
      # Вычислим формулу
      variants_count += math.factorial(n)/math.factorial(m)/math.factorial(n-m)

    print(int(variants_count))

  except ValueError: print('Не буду считать такие большие числа!')


for i in range(7):
  print('\n\nРешение задания № 9')
  run_z9()
````

# **Задание 10**

Напишите функцию, чтобы найти самый длинный префикс в листе, составленном из строк.

Если общего префикса нет - верните пустую строку.



```
Input 1: strs = ["flower","flow","flight"]
Output 1: "fl"

Input 2: strs = ["dog","racecar","car"]
Output 2: ""
Explanation: There is no common prefix among the input strings.
```



````python
#Решение задания 10
#YOUR CODE HERE

def run_z10():
  words = list(input('Введите слова через пробел: ').split())

  prefix = ''
  i = 1
  isdo = True

  while isdo:
    try:
      # Новым кандидатом в перфикс становятся
      # первые i символов первого слова из списка
      prefix_candidate = words[0][:i]
      for word in words:
        if prefix_candidate != word[:i]:
          # Если хоть в каком-то слове первые i символов не совпали
          # с кандидатом, заканчиваем все последующие проверки
          isdo = False
          break
      if isdo:
        # На этом этапе кандидат совпал с первыми i символами
        # во всех словах, значит кандидат становится итоговым перфиксом
        prefix = prefix_candidate
        i += 1 # А число символов для проверки увеличивается
    except:
      # Сюда мы попадаем на первом же слове, в котором меньше символов, чем
      # мы хотим проверять на данном этапе, а значит завершаем все проверки!
      break

  print('\nОбщий префикс в списке слов:')
  print(words)
  print('\nТакой: ', '"', prefix, '"', sep='')
  if not prefix: print('\nЧто означает, что общий префикс отсутствует.')


for i in range(4):
  print('\n\nРешение задания № 10')
  run_z10()
````
